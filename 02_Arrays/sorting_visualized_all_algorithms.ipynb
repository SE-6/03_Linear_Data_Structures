```python
import time

DELAY = 0.4  # animation speed
step_counter = 0


def reset_steps():
    global step_counter
    step_counter = 0


def show(arr, msg=""):
    global step_counter
    step_counter += 1
    print(f"\033[96mStep {step_counter}:\033[0m {arr}  \033[93m{msg}\033[0m")
    time.sleep(DELAY)
```

## ðŸ«§ Bubble Sort


```python
def bubble_sort(arr):
    reset_steps()
    n = len(arr)

    # Outer loop: number of passes
    for i in range(n):
        for j in range(0, n - i - 1):
            show(arr, f"compare {arr[j]} and {arr[j+1]}")
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                show(arr, "swap")
    print("\033[92mBubble Sort finished\033[0m")
    return arr


bubble_sort([5, 3, 8, 1, 4])
```

## ðŸ“Œ Selection Sort


```python
def selection_sort(arr):
    reset_steps()
    n = len(arr)

    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            show(arr, f"compare {arr[min_index]} and {arr[j]}")
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
        show(arr, "swap min into position")

    print("\033[92mSelection Sort finished\033[0m")
    return arr


selection_sort([5, 3, 8, 1, 4])
```

## ðŸƒ Insertion Sort


```python
def insertion_sort(arr):
    reset_steps()
    for i in range(1, len(arr)):
        current = arr[i]
        j = i - 1
        show(arr, f"take {current}")
        while j >= 0 and arr[j] > current:
            arr[j + 1] = arr[j]
            show(arr, f"shift {arr[j]}")
            j -= 1
        arr[j + 1] = current
        show(arr, f"insert {current}")

    print("\033[92mInsertion Sort finished\033[0m")
    return arr


insertion_sort([5, 3, 8, 1, 4])
```

## âš¡ Quick Sort


```python
def quick_sort(arr, left, right):
    if left < right:
        pivot = arr[right]
        i = left - 1
        show(arr, f"pivot = {pivot}")
        for j in range(left, right):
            if arr[j] <= pivot:
                i += 1
                arr[i], arr[j] = arr[j], arr[i]
                show(arr, "swap")
        arr[i + 1], arr[right] = arr[right], arr[i + 1]
        show(arr, "pivot placed")
        quick_sort(arr, left, i)
        quick_sort(arr, i + 2, right)
    return arr


reset_steps()
quick_sort([5, 3, 8, 1, 4], 0, 4)
print("\033[92mQuick Sort finished\033[0m")
```

## ðŸ”¢ Counting Sort


```python
def counting_sort(arr):
    reset_steps()
    max_val = max(arr)
    count = [0] * (max_val + 1)

    for num in arr:
        count[num] += 1
        show(count, f"count {num}")

    sorted_arr = []
    for value in range(len(count)):
        for _ in range(count[value]):
            sorted_arr.append(value)
            show(sorted_arr, f"append {value}")

    print("\033[92mCounting Sort finished\033[0m")
    return sorted_arr


counting_sort([5, 3, 8, 1, 4])
```

## ðŸ”¢ Radix Sort


```python
def radix_sort(arr):
    reset_steps()
    max_num = max(arr)
    exp = 1

    while max_num // exp > 0:
        show(arr, f"sorting by digit {exp}")
        counting_sort_digit(arr, exp)
        exp *= 10

    print("\033[92mRadix Sort finished\033[0m")
    return arr


def counting_sort_digit(arr, exp):
    n = len(arr)
    output = [0] * n
    count = [0] * 10

    for num in arr:
        digit = (num // exp) % 10
        count[digit] += 1

    for i in range(1, 10):
        count[i] += count[i - 1]

    for i in range(n - 1, -1, -1):
        digit = (arr[i] // exp) % 10
        output[count[digit] - 1] = arr[i]
        count[digit] -= 1

    for i in range(n):
        arr[i] = output[i]


radix_sort([170, 45, 75, 90])
```

## ðŸ”€ Merge Sort


```python
def merge_sort(arr):
    show(arr, "split")
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    merged = merge(left, right)
    show(merged, "merge")
    return merged


def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result


reset_steps()
merge_sort([5, 3, 8, 1, 4])
print("\033[92mMerge Sort finished\033[0m")
```
