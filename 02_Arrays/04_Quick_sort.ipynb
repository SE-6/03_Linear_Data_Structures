```python
def partition(arr, left, right):
    pivot_value = arr[right]
    i = left - 1
    for j in range(left, right):
        if arr[j] <= pivot_value:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]

    arr[i + 1], arr[right] = arr[right], arr[i + 1]
    return i + 1  # position of pivot


def quick_sort(arr, left, right):
    if left < right:
        # partion
        pivot_index = partition(arr, left, right)
        # quick sort all partitions
        quick_sort(arr, left, pivot_index - 1)
        quick_sort(arr, pivot_index + 1, right)

    # Base Case: left became as big right
    return arr


nums = [5, 2, 8, 5, 0, 23, 76, 34, 56, 23, 12]

quick_sort(nums, 0, len(nums) - 1)
```

```python
import random


def partition(arr, left, right):
    # For a random pivot, you can select an index between left and right
    # then swap the rightmost value
    pivot_index = random.randint(left, right)
    arr[pivot_index], arr[right] = arr[right], arr[pivot_index]

    # The rest of the logic can stay the same, but the algorithm should be faster
    pivot_value = arr[right]
    i = left - 1
    for j in range(left, right):
        if arr[j] <= pivot_value:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]

    arr[i + 1], arr[right] = arr[right], arr[i + 1]
    return i + 1


def quick_sort(arr, left, right):
    if left < right:
        pivot_index = partition(arr, left, right)
        quick_sort(arr, left, pivot_index - 1)
        quick_sort(arr, pivot_index + 1, right)
    return arr


nums = [5, 2, 8, 5, 0, 23, 76, 34, 56, 23, 12]
print(quick_sort(nums, 0, len(nums) - 1))
```
