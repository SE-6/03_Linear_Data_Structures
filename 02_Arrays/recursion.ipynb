```python
def factorial(num):
    total = 1
    count = num
    while count > 0:
        total *= count
        count -= 1

    return total


print(factorial(10))


# >>> factorial(5)
# 5 * 4 * 3 * 2 * 1 -> 120
```

```python
def factorial(num):
    #  Base Case: simplest version of the problem - when to stop recursion
    if num <= 1:
        return 1
    else:
        #  Recursive case
        return num * factorial(num - 1)


print(factorial(5))
```
