```python
print("Linear Search")
```

```python
def linear_search(arr, target):
    for index, value in enumerate(arr):
        if value == target:
            return index
    return -1
```

```python
numbers = [6, 2, 9, 4, 10, 7, 5, 1, 3, 8]
target = 5  # at which index is the first `5`?

found_ind = linear_search(numbers, target)

print(found_ind)
```

```python
print(numbers.index(target))
```

```python
best_case_target = 6
print(linear_search(numbers, best_case_target))  # O(1)

worst_target = 8
print(linear_search(numbers, worst_target))  # O(n) -> traverse whole list

average_target = 5
print(linear_search(numbers, average_target))  # O(n)
```
