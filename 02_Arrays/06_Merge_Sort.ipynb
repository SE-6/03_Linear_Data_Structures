```python
def merge(arr, left, mid, right):
    n1 = mid - left + 1
    n2 = right - mid
    # L = [0] * n1
    # R = [0] * n2
    # for i in range(n1):
    #     L[i] = arr[left + 1]
    # for i in range(n2):
    #     R[i] = arr[mid + 1 + i]
    L = arr[left : mid + 1]
    R = arr[mid + 1 : right + 1]  # Space complexity O(n) # needs to copy whole data

    i = 0
    j = 0
    k = left

    while i < n1 and j < n2:
        if L[i] <= R[j]:
            arr[k] = L[i]
            i += 1
        else:
            arr[k] = R[j]
            j += 1
        k += 1

    # Copy leftovers
    while i < n1:
        arr[k] = L[i]
        k += 1
        i += 1
    while j < n2:
        arr[k] = R[j]
        k += 1
        j += 1


def merge_sort(arr, left, right):  # O(n log n) -> Time complexity
    if left < right:
        mid = (right + left) // 2
        print(
            f"Split: {arr[left:right+1]} into {arr[left:mid+1]} and {arr[mid+1:right+1]}"
        )
        merge_sort(arr, left, mid)  # O(log n)
        merge_sort(arr, mid + 1, right)
        # Merge parts back together
        print(f"Merging: {arr[left:mid+1]} + {arr[mid+1:right+1]}")
        merge(arr, left, mid, right)  # O(n)
        print(f"Result:  {arr[left:right+1]}")
    return arr
```

```python
nums = [123, 345, 321, 645]

print(merge_sort(nums, 0, len(nums) - 1))

sorted(nums)
# Timsort -> combination of insertion and merge sort
# Powersort -> current default in Python
```
