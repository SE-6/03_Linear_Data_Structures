```python
some_list = [13, 123, 1, 4, 2, 8, 54, 3, 7, 4, 5]  # average case

already_sorted = [1, 2, 3, 4, 5]  # O(n)  - best case

worst_case = [5, 4, 3, 2, 1]  #  O(n²)


def bubble_sort(arr):
    n = len(arr)  # O(1)

    for i in range(n - 1):  # O(n)
        swapped = False  # O(1) - When no swap happened during the outer loop, the array is sorted already
        for j in range(0, n - 1 - i):  # O(n²) <--- dominant loop
            curr_val = arr[j]  # O(1)
            next_val = arr[j + 1]  # O(1)
            if curr_val > next_val:  # O(1)
                # swap
                arr[j], arr[j + 1] = arr[j + 1], arr[j]  # O(1)
                swapped = True  # O(1)
        if not swapped:  # O(1)
            break  # O(1)
    return arr


print(
    "before sort: ", some_list
)  # this is in-place sorting. It's more efficient that creating a copy first, but the original order is lost.
bubble_sort(some_list)
print("after sort: ", some_list)
```
