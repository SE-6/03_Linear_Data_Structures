```python
def selection_sort(arr):
    n = len(arr)  # O(1) - Constant Time
    for i in range(n - 1):  # O(n) - Linear Time
        min_index = i
        for j in range(i + 1, n):  # O(n²) - Quadratic <-- Dominant
            if arr[j] < arr[min_index]:
                min_index = j
        if min_index != i:
            arr[i], arr[min_index] = arr[min_index], arr[i]

    return arr


# O(n²)
```

```python
average_list = [7, 3, 6, 0, 8, 1, 6, 8, 3, 5, 8, 8, 8, 5, 5]

print(average_list)
selection_sort(average_list)
print(average_list)
```
