```python
class Stack:
    def __init__(self) -> None:
        self.stack = []

    def push(self, item):
        self.stack.append(item)

    def pop(self):
        if len(self.stack) == 0:
            return None
        return self.stack.pop()

    def size(self):
        return len(self.stack)

    def peek(self):
        return self.stack[-1]
```

```python
my_stack = Stack()
my_stack.push(123)
my_stack.push(123)
my_stack.push(123)
my_stack.push(123)
my_stack.push("top")
print(my_stack.size())
last = my_stack.pop()
print(last)
print(my_stack.size())

my_stack.pop()
my_stack.pop()
my_stack.pop()
my_stack.pop()
my_stack.pop()
my_stack.pop()
my_stack.pop()
my_stack.pop()
my_stack.pop()
last = my_stack.pop()
print(last)
```
