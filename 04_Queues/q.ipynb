```python
class Queue:
    def __init__(self) -> None:
        self.q = []

    def enqueue(self, item):  # O(1)
        self.q.append(item)

    def dequeue(self):  # O(n)
        if len(self.q) == 0:
            return None
        return self.q.pop(0)

    def size(self):  # O(1)
        return len(self.q)

    def peek(self):  # O(1)
        return self.q[0]
```

```python
my_q = Queue()
my_q.enqueue("First frame")
my_q.enqueue("Second frame")
my_q.enqueue("Third frame")

print(my_q.peek())
print(my_q.size())
to_process = my_q.dequeue()
print(to_process)
print(my_q.size())
```

```python
class Queue:
    def __init__(self) -> None:
        self.q = []
        self.head = 0

    def enqueue(self, item):  # O(1)
        self.q.append(item)

    def dequeue(self):  # O(1)
        if len(self.q) == 0:
            return None
        dequeued_val = self.q[self.head]
        self.head += 1
        return dequeued_val

    def size(self):  # O(1)
        return len(self.q)

    def peek(self):  # O(1)
        return self.q[0]
```
