```python
# This function is doing, what it's supposed to: finding the largest number in a list.
# But when the list grows, the time it takes to run this function grows a lot more.
# len(arr) == 4 -> Time: 16
# len(arr) == 5 -> Time: 25
# It's because of the nested loop in the function. It needs to run over the list n * n times
# O(n²)


def find_largest(arr):
    n = len(arr)  # 1
    for i in range(n):  # n
        is_largest = True  # n * 1
        for j in range(n):  # n
            if arr[j] > arr[i]:  # n * 1
                is_largest = False  # 1
                break
        if is_largest:  # 1
            return arr[i]


#  3 + 2n + n²
# 10
#  3 + 20 + 100
# 100
# 3 + 200 + 10000

# -> n²

numbers = [3, 2, 5, 4, 6, 7, 8, 5, 99]  # len(numbers) ** 2 -> O(n²)
largest = find_largest(numbers)
print("Largest number: ", largest)
```

```python
#  Here is a faster function. Same input, same output, but way faster.
# It needs to run only once over the list of numbers.
# len(arr) == 4 -> Time: 4
# len(arr) == 5 -> Time: 5
# Most importantly, when the size of the input grows, the time grows only in a linear fashion.
# O(n)


def find_largest(arr):
    max_num = arr[0]  # O(1)
    for i in range(
        1, len(arr)
    ):  # O(n - 1) # the -1 doesn't make a difference on large numbers
        if arr[i] > max_num:  # O(1)
            max_num = arr[i]  # O(1)
    return max_num


numbers = [3, 1, 4, 1, 5, 9, 4, 5, 7, 8, 9, 0]


largest = find_largest(numbers)
print("Largest number: ", largest)
```
